```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the updated dataset.
Insu = pd.read_csv("Medical_cost/updated_insurance.csv")
```

```python
plt.hist(Insu["age"])
plt.title("Age Distribution")
plt.show()

"""
Insight Paragraph:
The age histogram shows that ages are spread fairly evenly across the dataset.
There is no single age group that dominates the data, meaning the dataset includes
a wide range of individuals. Because the ages are well distributed, it allows us
to observe how insurance charges may change across different age groups without
strong bias toward a specific age range.
"""
```

```python
plt.hist(Insu["bmi"])
plt.title("BMI Distribution")
plt.show()

"""
Insight Paragraph:
The BMI histogram shows that most BMI values are concentrated around the middle
range, with fewer people having extremely low or extremely high BMI values.
This indicates that the majority of individuals in the dataset fall into moderate
BMI categories. While extreme BMI values exist, they are less common in the data.
"""
```

```python
plt.hist(Insu["charges"])
plt.title("Charges Distribution")
plt.show()

"""
Insight Paragraph:
The charges histogram is right-skewed, meaning most people have relatively low
insurance costs while a smaller number of individuals have very high charges.
This suggests that certain factors in the dataset may significantly increase
insurance costs for some individuals, creating the long tail on the right side
of the distribution.
"""
```

```python
sns.boxplot(x=Insu["smoker"], y=Insu["charges"])
plt.show()

"""
Insight Paragraph:
The boxplot comparing smokers and non-smokers shows a clear difference in
insurance charges between the two groups. Smokers have a much higher median
charge and a wider range of costs compared to non-smokers. This indicates that
smoking status strongly affects insurance expenses and is likely one of the
most influential variables in the dataset.
"""
```

```python
sns.boxplot(x=Insu["sex"], y=Insu["charges"])
plt.show()

"""
Insight Paragraph:
The boxplot comparing sex and insurance charges shows that males and females
have very similar distributions of charges. The median values and overall
ranges are close for both groups. This suggests that gender does not have a
major impact on insurance costs in this dataset compared to other variables.
"""
```

```python
corr = Insu.corr(numeric_only=True)

sns.heatmap(corr)
plt.title("Correlation Heatmap of Insurance Features")
plt.show()

"""
Insight Paragraph:
The heatmap displays the correlation between the numerical features in the
dataset. Some variables show stronger relationships with charges than others,
while many features have weak correlations. This visualization helps identify
which features may have more influence on insurance costs and which ones are
less important for explaining the target variable.
"""
```

```python
sns.scatterplot(x=Insu["bmi"], y=Insu["charges"], hue=Insu["smoker"])
plt.show()

"""
Insight Paragraph:
The scatter plot shows the relationship between BMI and insurance charges,
while also distinguishing smokers and non-smokers by color. Smokers tend to
have much higher charges than non-smokers across most BMI values. In addition,
charges for smokers appear to increase more as BMI increases, suggesting that
smoking combined with higher BMI may contribute to higher insurance costs.
"""
```

```python
Insu.groupby("smoker")["charges"].mean()

"""
Insight Paragraph:
The groupby analysis shows the average insurance charges for smokers and
non-smokers. Smokers have a much higher average charge compared to
non-smokers, indicating that smoking is strongly associated with higher
medical insurance costs. This confirms the pattern already observed in
the earlier visualizations.
"""
```
