```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

Insu = pd.read_csv("updated_insurance.csv")
```

```python
# Computing the mean and standard deviation of charges.
mean = np.mean(Insu["charges"])
std = np.std(Insu["charges"])

# Print the calculated mean and standard deviation of the dataset feature
print(f"""Mean:{mean}
Standard deviation:{std}""")
```

```python
# Standardising BMI column by hand using broadcasting: z = (X - mean) / std.
mean = np.mean(Insu["bmi"])
std = np.std(Insu["bmi"])

z = (Insu["bmi"] - mean) / std

# Display the first five values of the Z-score calculation done using broadcasting
print(f"Broadcasting:{z.head()}")
```

```python
# BMI Standard Scaling.
stan = StandardScaler()
bmi = stan.fit_transform(Insu[['bmi']])

# Display the first five values after scaling the BMI feature using StandardScaler
print(f"StandardScaler:{bmi[:5]}")

# Both methods produce very similar results, confirming the manual standardization works.
```

```python
# Computing cosine similarity between feature vectors of highest and lowest charges

# Find rows
Highs = Insu['charges'].idxmax()
Lows = Insu['charges'].idxmin()

# Keep only numeric columns
num = Insu.select_dtypes(include=np.number)

# Extract vectors
Highs = num.loc[Highs]
Lows = num.loc[Lows]

# Compute cosine similarity
cosine = np.dot(Highs, Lows) / (np.linalg.norm(Highs) * np.linalg.norm(Lows))

# Print the cosine similarity between the selected feature vectors
print(cosine)
```

```python
# Initialize a counter for the total number of smokers
smokertotal = 0

# Initialize a counter for smokers whose charges exceed the threshold
smoker_with_charges_above_threshold = 0

# Loop through every row in the dataset
for i in range(len(Insu)):

    # Check if the person in the current row is a smoker
    if Insu["smoker"].iloc[i] == 1:

        # Increase the total smoker count
        smokertotal += 1

        # Check if this smoker's charges are above the threshold (20000)
        if Insu["charges"].iloc[i] > 20000:

            # Increase the count of smokers with high charges
            smoker_with_charges_above_threshold += 1

# Calculate the probability that a smoker has charges above the threshold
probability = smoker_with_charges_above_threshold / smokertotal

# Print how many smokers exceed the threshold out of the total smokers
print(f"{smoker_with_charges_above_threshold} of {smokertotal} smokers have charges above 20000")

# Print the calculated probability
print(f"Probability: {probability}")
```
